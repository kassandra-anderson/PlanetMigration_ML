```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

%matplotlib widget
```

```python
# The purpose of this notebook is to visually inspect the distributions of various input features and decide which features to log transform

seed = 0

path_to_data = "/Users/kassand/astro/PlanetMigration_ML/training_data/"
df = pd.read_csv(path_to_data + "alldata_additional_features.csv")
df = shuffle(df,random_state=seed)
```

```python
def plot_feat_logfeat(feature, name):
    
    ''' Takes a specified feature and plots both the feature and the log of the feature '''
    
    plt.figure(figsize = (8, 4))
    plt.subplot(1,2,1)
    plt.hist(feature)
    plt.xlabel(name)
    plt.subplot(1,2,2)
    plt.hist(np.log10(feature))
    plt.xlabel('log ' + name)
    plt.show()
```

```python
plot_feat_logfeat(df["eps_gr_input"],"epsGR")
plot_feat_logfeat(df["eps_oct_input"],"epsOct")
plot_feat_logfeat(df["eps_tide1_input"],"epsTide1")
plot_feat_logfeat(df["eps_tide2_input"],"epsTide2")
plot_feat_logfeat(df["eps_rot1_input"],"epsRot1")
plot_feat_logfeat(df["eps_rot2_input"],"epsRot2")

plot_feat_logfeat(df["L1x_input"],"L1x")
plot_feat_logfeat(df["L1y_input"],"L1y")
plot_feat_logfeat(df["L1z_input"],"L1")

plot_feat_logfeat(df["L2x_input"],"L2x")
plot_feat_logfeat(df["L2y_input"],"L2y")
plot_feat_logfeat(df["L2z_input"],"L2z")

plot_feat_logfeat(df["e1x_input"],"e1x")
plot_feat_logfeat(df["e1y_input"],"e1y")
plot_feat_logfeat(df["e1z_input"],"e1z")

plot_feat_logfeat(df["e2x_input"],"e2x")
plot_feat_logfeat(df["e2y_input"],"e2y")
plot_feat_logfeat(df["e2z_input"],"e2z")

plot_feat_logfeat(df["cos_Imut_input"],"cosI")
plot_feat_logfeat(df["aF_max_input"],"aFmax")
plot_feat_logfeat(df["peri_max_input"],"perimax")
plot_feat_logfeat(df["peri_max_input"]/df["Rtide_input"],"perimax_over_Rtide")

plot_feat_logfeat(df["aF_max7_input"],"aFmax7")

plot_feat_logfeat(df["Rtide_input"],"Rtide")
```
